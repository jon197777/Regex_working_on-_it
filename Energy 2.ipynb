```python
import requests
import pdfplumber
import pandas as pd
import re

```

```python
def download_file(url):
  local_filename = url.split('/')[-1]

  with requests.get(url) as r:
    with open(local_filename, 'wb') as f:
      f.write(r.content)

  return local_filename

ap_url = 'https://qnadailyreport.blob.core.windows.net/qnadailyreportxml/Written-Questions-Answers-Statements-Daily-Report-Commons-2022-11-25.pdf'

  

```

```python
ap = download_file(ap_url)
```

```python
with pdfplumber.open(ap) as pdf:
  page = pdf.pages[4]
  text = page.extract_text()
```

```python
print(text)
```

# Main Title

```python
question_title = re.compile(r'  [A-Z]+,')
```

```python
for line in text.split('\n'):
  if question_title.match(line):
    print(line)
```

#Sub Title

```python
question_sub_title = re.compile(r'  [A-Z][a-z]{1,}|: [A-Z][a-z]{1,}')
```

```python
for line in text.split('\n'):
  if question_sub_title.match(line):
    print(line)
```

# Name to Question Posed

```python
question_name = re.compile('[A-Z][a-z]{1,} [A-Z][a-z]{1,}:   ')
```

```python
for line in text.split('\n'):
  if question_name.match(line):
    print(line)
```

# Name to Answers Given

```python
answer_name = re.compile('[A-Z][a-z]{1,} [A-Z][a-z]{1,}: $')
```

```python
for line in text.split('\n'):
  if answer_name.match(line):
    print(line)
```

# Unique Question ID

```python
question_ID = re.compile('[8    8]')
```

```python
for line in text.split('\n'):
  if question_ID.match(line):
    print(line)
    
```
